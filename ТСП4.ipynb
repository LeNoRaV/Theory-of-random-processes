```python
import pandas as pd
import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
import seaborn as sb
import random
from prettytable import PrettyTable

np.set_printoptions(precision=3, suppress=True)

def print_table(th, td):
    table = PrettyTable(th)
    columns = len(th)
    td_data = td.tolist()[:]
    while td_data:
        table.add_row(td_data[:columns])
        td_data = td_data[columns:]
    print(table)
```

```python
a=0
b=5
h=0.05
k=100
n=160
m=[]
for i in range (0, k+1):
  t=i*h
  m.append(2-math.exp(-t))
  #m.append(3-2*math.exp(-t))
print('m =', m)
print('t =', t)
print(len(m))
```

```python
S=[]
for i in range (0,k+1):
  buf=[]
  for j in range (0,k+1):
    t=abs(i-j)*h
    buf.append(3/2*math.exp(-2*abs(t)))
  S.append(buf)

print('\n'.join('\t'.join(map(str, row)) for row in S))
```

```python
L=np.linalg.cholesky(S)
R=[]
for i in range(k+1):
  buf=[]
  for j in range(k+1):
    buf.append(round(L[i][j],4))
  R.append(buf)
print('\n'.join('\t'.join(map(str, row)) for row in R))
```

```python
x=[]
for i in range(n):
  buf=[]
  for j in range(k+1):
    buf.append(np.random.randn())
  x.append(buf)

np.savetxt('test.txt', x, fmt='%f')

e = np.loadtxt('test.txt', dtype=float)

print('\n'.join('\t'.join(map(str, row)) for row in e))
```

```python
eta=[]
for i in range (n):
  eta.append(np.dot(L,e[i]))

N=[]
for i in range(n):
  for j in range(k+1):
    N.append(round(eta[i][j],4))

print( N)
```

```python
ksi=[]
for i in range(n):
  buf=[]
  for j in range(k+1):
    buf.append(eta[i][j]+m[j])
  ksi.append(buf)

R=[]
for i in range(n):
  buf=[]
  for j in range(k+1):
    buf.append(round(ksi[i][j],4))
  R.append(buf)
  print(' ',i,')',R[i])
print(len(eta[1]))
```

```python
print("Приведем графическую иллюстрацию последовательного критерия\n")
xx = np.arange (a, b+h, h)
sb.set_style("whitegrid")
plt.plot(xx, ksi[0], color='red',label='ksi[1]')
plt.plot(xx, ksi[9], color='green',label='ksi[10]')
plt.plot(xx, ksi[n-1], color='blue',label='ksi[n]')
plt.plot(xx, m, color='black',label='m')
plt.xlabel("t")
plt.legend()
plt.show()
```

```python
l=math.trunc(1 + math.log2(n))
print(l)
Max = np.amax(ksi[0])
Min = np.amin(ksi[0])
w = Max - Min
h = w / l
print(h)
print(w)
```

```python
print("Число интервалов l = %d\nШаг(ширина интервалов группировки) h = %f" %(l, h))
x=ksi[0]
hist, bin_edges = np.histogram(x, l)
bin_edges_average = np.zeros(bin_edges[:l].size)
for i in range (l):
    bin_edges_average[i] = (bin_edges[i] + bin_edges[i + 1]) / 2
f1 = bin_edges_average
f2 = hist

print("\n")
th = ["Cредние значения каждого из  интервалов", "Частоты попадания элементов в каждый из интервалов"]
td = np.array(f1[0])
for i in range(len(f1)-1):
    td = np.append(td, f2[i])
    td = np.append(td, f1[i+1])
td = np.append(td, f2[6])
print_table(th, td)

p = hist / n

Int = f1
intl = Int - h / 2
intl = np.append(intl, Max)

print("\n")
th = ["Интервал", "Середина интервала int", "Частота f2", "Относительная частота p"]
td = []
for i in range(l):
    td.append("["+str(round(intl[i], 4))+","+str(round(intl[i + 1], 4))+")")
    td.append(round(Int[i], 4))
    td.append(f2[i])
    td.append(round(p[i], 4))
print_table(th, np.array(td))
```
